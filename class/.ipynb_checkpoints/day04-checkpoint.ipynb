```python
p_ptratio = len(ptratio)/len(df)
```

```python
#1/(n-ddof), pandas ddof default=1
print(df['RM'].var(), df['RM'].var(ddof=0), df['RM'].var(ddof=1))
print(df['RM'].std(), df['RM'].std(ddof=0), df['RM'].std(ddof=1))
```

```python
#numpy ddof default=0
print(np.var(df['RM']), np.var(df['RM'], ddof=0), np.var(df['RM'], ddof=1))
print(np.std(df['RM']), np.std(df['RM'], ddof=0), np.std(df['RM'], ddof=1))
```

```python
X = df[['CRIM', 'NOX', 'RM', 'AGE', 'TAX']]
```

```python
#numpy는 rowvar 0 해야 열을 변수로
cov_x = np.cov(X, rowvar=0)
#cov_x = np.cov(X.T)
print(cov_x)
```

```python
#분산 공분산 동시
cov_x = X.cov()
cov_x
```

```python
import seaborn as sns
corr_x = X.corr()
sns.heatmap(corr_x, annot=True, cmap='Reds')
```
