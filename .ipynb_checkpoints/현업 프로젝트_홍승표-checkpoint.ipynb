### 1. 각 wafer 별 x, y 매핑 이미지 데이터 생성

### wafer 제품별로 측정값의 상대적인 분포에 따라 불량이 판단되므로 train 데이터 전체를 기준으로 수행하지 않고 각 wafer 단위 min max value로 scaling을 수행합니다.

### gray scale 이미지 표현을 위해 0~255 사이 min max scaling 후 값 반전하여 gray scale 값을 생성합니다.

```python
def scale_to_gray(data):
    min_val = np.min(data)
    max_val = np.max(data)
    scaled_data = 255 - (255 * (data - min_val) / (max_val - min_val))
    scaled_data_int = scaled_data.astype(np.uint8)
    return scaled_data_int
```

```python
fig, axes = plt.subplots(nrows=1, ncols=len(raw_col), figsize=(20, 6))
for i, col in enumerate(raw_col):
    map = np.zeros([116, 194])
    for index, row in df_wf.iterrows():
        map[row['x']][row['y']] = row[col]
    gray_map = scale_to_gray(map)
    #plt.imshow(gray_map, cmap='gray', aspect='auto')
    #plt.savefig(f"image_{col}.png")

    axes[i].imshow(gray_map, cmap='gray', aspect='auto')
    axes[i].set_title(f'Gray Scale Map for {col}', fontsize=14)
    axes[i].set_xlabel('Y')
    axes[i].set_ylabel('X')

plt.tight_layout()
plt.show()
```

### 생성한 map 데이터를 resize하고 학습 feature 벡터를 생성합니다.
### 학습 데이터는 원본 데이터 기준으로 웨이퍼의 상단/하단부에 해당하는 각 20행 데이터를, resize 기준으로 상하단 5개 행을 추출하여 불량 판단에 사용합니다.

```python
gray_map.shape
```

```python
import cv2
fig, axes = plt.subplots(nrows=1, ncols=len(raw_col), figsize=(20, 6))
for i, col in enumerate(raw_col):
    map = np.zeros([116, 194])
    for index, row in df_wf.iterrows():
        map[row['x']][row['y']] = row[col]
    gray_map = scale_to_gray(map)
    #plt.imshow(gray_map, cmap='gray', aspect='auto')
    #plt.savefig(f"image_{col}.png")
    resized_map = cv2.resize(gray_map, (15, 25))
    axes[i].imshow(resized_map, cmap='gray', aspect='auto')
    axes[i].set_title(f'Gray Scale Map for {col}', fontsize=14)
    axes[i].set_xlabel('Y')
    axes[i].set_ylabel('X')

plt.tight_layout()
plt.show()

```

```python
train_x_resize = []
train_y_resize = []
for wf_idx in range(1,101):
    df_wf = df[df['wafer']==wf_idx]
    for col in raw_col:
        map = np.zeros([116, 194])
        for index, row in df_wf.iterrows():
            map[row['x']][row['y']] = row[col]
            
        gray_map = scale_to_gray(map)
        resized_map = cv2.resize(gray_map, (15, 25))
        
        top_rows = resized_map[:5, :]   # 위에서부터 5행
        bottom_rows = resized_map[-5:, :]  # 아래에서부터 5행
        
        top_bottom = np.concatenate((top_rows, bottom_rows))
        wf_vector = top_bottom.flatten()
        
        train_x_resize.append(wf_vector)
        train_y_resize.append(col)
```

### train / test 데이터를 분할하고 Logistic Regression 모델을 통해 학습 및 평가를 수행합니다.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(train_x_resize, train_y_resize, test_size=0.2, random_state=42)

model = LogisticRegression(penalty='l2')
model.fit(X_train, y_train)
y_test_pred = model.predict(X_test)
test_accuracy = accuracy_score(y_test, y_test_pred)

print(f"Test Accuracy: {test_accuracy:.4f}")
```

### 학습 모델은 pytorch를 기반으로 파라미터 수에 비해 성능이 좋은 EfficientNet을 활용하여 학습을 진행하고 모델 성능을 평가합니다.

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset, random_split
from torchvision import models, transforms
from sklearn.preprocessing import LabelEncoder
```

### 모델 input에 맞는 전처리 과정을 수행하고 학습을 진행합니다.

```python
transform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])
```

```python
class CustomImageDataset(Dataset):
    def __init__(self, images, labels, transform=None):
        self.images = images
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        image = self.images[idx]
        label = self.labels[idx]

        if len(image.shape) == 2:
            image = np.repeat(image[:, :, np.newaxis], 3, axis=2)

        if self.transform:
            image = self.transform(image)

        return image, label
```

### 모델은 Efficientnet b0을 기준으로 학습을 진행하고 평가를 수행하려고 합니다.

```python
class EfficientNetClassifier(nn.Module):
    def __init__(self, num_classes):
        super(EfficientNetClassifier, self).__init__()
        self.model = models.efficientnet_b0(pretrained=True)
        self.model.classifier[1] = nn.Linear(self.model.classifier[1].in_features, num_classes)

    def forward(self, x):
        return self.model(x)
```

### 라벨을 인코딩하여 지도학습을 위한 입력으로 사용합니다.

```python
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)
```

### 입력 tensor와 dataset, dataloader를 생성하고 8:2 비율인 train set을 다시 8:2로 분할하여 validation set을 생성합니다.

```python
y_train_tensor = torch.tensor(y_train_encoded, dtype=torch.long)
y_test_tensor = torch.tensor(y_test_encoded, dtype=torch.long)
```

```python
train_dataset = CustomImageDataset(X_train, y_train_tensor, transform=transform)
test_dataset = CustomImageDataset(X_test, y_test_tensor, transform=transform)
```

```python
train_size = int(0.8 * len(train_dataset))
val_size = len(train_dataset) - train_size
train_dataset, val_dataset = random_split(train_dataset, [train_size, val_size])
```

```python
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)
```

### 분류 class 개수와 모델을 정의하고 학습을 수행합니다.

```python
num_classes = 5
model = EfficientNetClassifier(num_classes)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

### 학습은 수행 시간을 고려하여 20epoch 수행합니다.

```python
num_epochs = 20
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)
```

### 학습 과정에서의 loss 변화를 result list에 저장합니다.

```python
train_loss_result = []
val_loss_result = []
```

```python

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * inputs.size(0)

    epoch_loss = running_loss / len(train_loader.dataset)
    
    # Validation phase
    model.eval()
    val_loss = 0.0
    correct = 0
    total = 0
    with torch.no_grad():
        for inputs, labels in val_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            val_loss += loss.item() * inputs.size(0)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    val_loss = val_loss / len(val_loader.dataset)
    val_accuracy = correct / total
    
    print(f'Epoch {epoch + 1}/{num_epochs}, Loss: {epoch_loss:.4f}, Val Loss: {val_loss:.4f}, Val Accuracy: {val_accuracy:.4f}')
    train_loss_result.append(epoch_loss)
    val_loss_result.append(val_loss)
```

```python
model.eval()
correct = 0
total = 0
with torch.no_grad():
    for inputs, labels in test_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        outputs = model(inputs)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

accuracy = correct / total
print(f'Test Accuracy: {accuracy:.4f}')
```

```python
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

y_train_tensor = torch.tensor(y_train_encoded, dtype=torch.long)
y_test_tensor = torch.tensor(y_test_encoded, dtype=torch.long)

train_dataset = CustomImageDataset(X_train, y_train_tensor, transform=transform)
test_dataset = CustomImageDataset(X_test, y_test_tensor, transform=transform)

train_size = int(0.1 * len(train_dataset))
val_size = len(train_dataset) - train_size
train_dataset, val_dataset = random_split(train_dataset, [train_size, val_size])

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)

num_classes = 5
model = EfficientNetClassifier(num_classes)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

num_epochs = 20
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

train_loss_result = []
val_loss_result = []
```

```python
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * inputs.size(0)

    epoch_loss = running_loss / len(train_loader.dataset)
    
    model.eval()
    val_loss = 0.0
    correct = 0
    total = 0
    with torch.no_grad():
        for inputs, labels in val_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            val_loss += loss.item() * inputs.size(0)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    val_loss = val_loss / len(val_loader.dataset)
    val_accuracy = correct / total
    
    print(f'Epoch {epoch + 1}/{num_epochs}, Loss: {epoch_loss:.4f}, Val Loss: {val_loss:.4f}, Val Accuracy: {val_accuracy:.4f}')
    train_loss_result.append(epoch_loss)
    val_loss_result.append(val_loss)
```

### 불분명한 불량 분류 문제는 주로 적은 cell의 불량에서 기인하는 경우가 많기 때문에 확률적으로 불량 값을 제거하여 각 불량 유형 간의 차이를 감소시킨 데이터를 활용합니다.

### => 불량의 강한 특징을 제거하기 위해 training 데이터는 0.95의 확률로, test 데이터는 0.97의 확률로 불량을 제거하여 미세한 불량에 대한 판단 문제로 축소합니다.

```python
from PIL import Image
def add_white_noise(img, probability=0.1):
    img = np.array(img)
    h, w = img.shape
    noise = np.random.rand(h, w)
    img[noise < probability] = 255  # 흰색으로 설정 (grayscale)
    image = Image.fromarray(img)
    image = np.array(image)
    return image
```

### 이전 과정과 동일하게 EfficientNet을 사용하여 학습을 진행합니다.

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset, random_split
from torchvision import models, transforms
from sklearn.preprocessing import LabelEncoder
```

```python
transform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])
```

```python
class CustomImageDataset(Dataset):
    def __init__(self, images, labels, transform=None):
        self.images = images
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        image = self.images[idx]
        label = self.labels[idx]

        if len(image.shape) == 2:  
            image = np.repeat(image[:, :, np.newaxis], 3, axis=2)

        if self.transform:
            image = self.transform(image)

        return image, label
```

```python
class EfficientNetClassifier(nn.Module):
    def __init__(self, num_classes):
        super(EfficientNetClassifier, self).__init__()
        self.model = models.efficientnet_b0(pretrained=True)
        self.model.classifier[1] = nn.Linear(self.model.classifier[1].in_features, num_classes)

    def forward(self, x):
        return self.model(x)
```

```python
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

y_train_tensor = torch.tensor(y_train_encoded, dtype=torch.long)
y_test_tensor = torch.tensor(y_test_encoded, dtype=torch.long)

train_dataset = CustomImageDataset(X_train, y_train_tensor, transform=transform)
test_dataset = CustomImageDataset(X_test, y_test_tensor, transform=transform)

train_size = int(0.8 * len(train_dataset))
val_size = len(train_dataset) - train_size
train_dataset, val_dataset = random_split(train_dataset, [train_size, val_size])

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)

num_classes = 5
model = EfficientNetClassifier(num_classes)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

num_epochs = 20
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

train_loss_result = []
val_loss_result = []
```

```python
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * inputs.size(0)

    epoch_loss = running_loss / len(train_loader.dataset)
    
    # Validation phase
    model.eval()
    val_loss = 0.0
    correct = 0
    total = 0
    with torch.no_grad():
        for inputs, labels in val_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            val_loss += loss.item() * inputs.size(0)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    val_loss = val_loss / len(val_loader.dataset)
    val_accuracy = correct / total
    
    print(f'Epoch {epoch + 1}/{num_epochs}, Loss: {epoch_loss:.4f}, Val Loss: {val_loss:.4f}, Val Accuracy: {val_accuracy:.4f}')
    train_loss_result.append(epoch_loss)
    val_loss_result.append(val_loss)
```

```python
ep = [i for i in range(1,21)]
plt.plot(ep, train_loss_result, marker='o', linestyle='-', color='b', label='train')
plt.plot(ep, val_loss_result, marker='o', linestyle='-', color='r', label='validation')
plt.title('Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
model.eval()
correct = 0
total = 0
with torch.no_grad():
    for inputs, labels in test_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        outputs = model(inputs)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

test_accuracy = correct / total
print(f'Test Accuracy: {test_accuracy}')
```

### 데이터를 생성하기 위해 GAN을 활용하여 Discriminator와 Generator 모델을 각각 생성하고 학습합니다.

```python
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Input, Concatenate, LeakyReLU
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from keras.initializers import RandomNormal
```

```python
latent_dim = 100
input_dim = X_train.shape[1]
num_classes = 5
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_train_one_hot = to_categorical(y_train_encoded, num_classes=num_classes)
```

### 모델은 입력보다 큰 노드수의 두 히든 레이어를 갖도록 generator와 discriminator를 생성해주고 활성화 함수로 relu와 분류를 위한 sigmoid 함수를 사용합니다.

```python
gen_input = Input(shape=(latent_dim,))
gen_label = Input(shape=(num_classes,))
gen_concat = Concatenate()([gen_input, gen_label])
gen = Dense(8192, activation='relu')(gen_concat)
gen = Dense(10240, activation='relu')(gen)
gen_output = Dense(input_dim, activation='sigmoid')(gen)
generator = Model([gen_input, gen_label], gen_output)

disc_input = Input(shape=(input_dim,))
disc_label = Input(shape=(num_classes,))
disc_concat = Concatenate()([disc_input, disc_label])
disc = Dense(10240, activation='relu')(disc_concat)
disc = Dense(8192, activation='relu')(disc)
disc_output = Dense(1, activation='sigmoid')(disc)
discriminator = Model([disc_input, disc_label], disc_output)
```

```python
discriminator.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['accuracy'])
discriminator.trainable = False
```

```python
gan_input = Input(shape=(latent_dim,))
gan_label = Input(shape=(num_classes,))
gan_gen_output = generator([gan_input, gan_label])
gan_disc_output = discriminator([gan_gen_output, gan_label])
gan = Model([gan_input, gan_label], gan_disc_output)
gan.compile(optimizer=Adam(), loss='binary_crossentropy')
```

```python
epochs = 100
batch_size = 32
for epoch in range(epochs):
    noise = np.random.randn(batch_size, latent_dim)
    sampled_labels = np.random.randint(0, num_classes, batch_size)
    sampled_labels_one_hot = to_categorical(sampled_labels, num_classes=num_classes)
    generated_data = generator.predict([noise, sampled_labels_one_hot])

    real_data = X_train[np.random.randint(0, X_train.shape[0], batch_size)]
    real_labels = y_train_one_hot[np.random.randint(0, y_train_one_hot.shape[0], batch_size)]

    # Discriminator 학습
    X = np.concatenate([real_data, generated_data])
    y = np.concatenate([np.ones((batch_size, 1)), np.zeros((batch_size, 1))])
    labels = np.concatenate([real_labels, sampled_labels_one_hot])
    d_loss = discriminator.train_on_batch([X, labels], y)

    # Generator 학습
    noise = np.random.randn(batch_size, latent_dim)
    y_gen = np.ones((batch_size, 1))
    g_loss = gan.train_on_batch([noise, sampled_labels_one_hot], y_gen)

    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Discriminator Loss: {d_loss[0]}, Generator Loss: {g_loss}")
```

### 학습 모델을 바탕으로 분류 모델에 사용할 데이터를 생성합니다.

```python
def generate_data(generator, latent_dim, class_label, num_samples=1):
    noise = np.random.randn(num_samples, latent_dim)
    label_one_hot = to_categorical([class_label] * num_samples, num_classes=num_classes)
    generated_data = generator.predict([noise, label_one_hot])
    return generated_data * 255.0
```

초기 정의한 문제는 웨이퍼 테스트 과정에서 발생하는 상하단 불량을 분류하는 문제였습니다. 이 문제는 웨이퍼의 map 데이터 중 국소 지역의 불량을 확인하는 과정이므로, 일부 영역만을 추출하여 분류하면 되는 문제로 축소하였습니다.

일부 영역만을 feature로 사용하기 때문에 머신러닝과 CNN 모두 활용할 수 있을 것 같다고 생각하였고 더 나은 모델을 찾아 적용하기 위해 세부 문제로 머신러닝 모델과 딥러닝 모델의 성능 비교 문제를 정의하였습니다. 그리고 추가적으로 데이터의 라벨링이나 일부 불량 데이터의 수집 자체가 어려울 경우도 고려하여 데이터 생성 문제도 함께 고려하였습니다.

초기 분류 결과는 사용한 데이터의 불량 분포 형태가 뚜렷하게 구분되어 머신러닝과 딥러닝 모델에서 모두 정확도 1.0의 결과를 얻을 수 있었습니다. 하지만 실제 테스트 결과로 발생하는 불량은 정상인지 불량인지조차 애매한 경우가 존재하여 이러한 문제를 해결하는 과정이 필요하다고 생각하였습니다. 때문에 데이터에 노이즈를 추가하거나 약한 불량의 데이터를 만드는 등의 작업을 진행해보았습니다. 이 과정에서 가우시안 노이즈 등의 augmentation보다 불량을 필터링하여 약한 불량을 만드는 방향이 실제 애매한 불량 데이터에 가깝다고 생각했고 이러한 데이터로 학습 데이터 세트를 구성하였습니다.

데이터를 보완한 결과 머신러닝 분류 모델에서 약 0.2~0.5의 정확도로 성능이 낮게 측정되었고 파라미터 튜닝과 앙상블을 적용하더라도 0.6 내외의 정확도로 만족스러운 성능은 얻지 못했습니다. 성능 비교를 위해 딥러닝 모델로 동일한 문제 해결 과정을 수행하였고, CNN 모델인 EfficientNet의 경우 노이즈 처리를 한 데이터에 대해서도 1.0의 정확도를 가지며 강한 성능을 보였습니다. 이 프로젝트 뿐만 아니라, 웨이퍼의 테스트 데이터는 웨이퍼의 셀 좌표와 함께 주어짐을 고려하면 다양한 test 데이터를 처리하는 방식에 이미지 처리를 통한 CNN 추론이 머신러닝보다 적합할 것이라는 결론을 얻을 수 있었습니다.

이 모델을 실제 불량에 대해 적용한다면, 실시간으로 발생하는 test 데이터로부터 불량 유형을 구분할 수 있을 것입니다. test 과정에서 기인된 불량이 검출되면 이를 바탕으로 해당 웨이퍼에 대한 재검사 알람 등으로 연계할 수 있고 효율적인 불량 대응 및 수율 개선이 가능할 것입니다. 더 나아가 높은 정확도가 확보되고 충분히 검증된다면 제조 시스템과 연계하여 자동 재검사 로직을 구현하는 방향으로도 적용해볼 수 있을 것 같습니다.

모델 성능을 개선하고 비교하는 과정을 진행하면서 모델 자체의 하이퍼파라미터 등을 조율하는 과정도 중요하지만, 사용하는 데이터에 적합한 특성을 가진 모델을 선정하고 활용하는 것이 굉장히 중요한 과정임을 깨닫게 되었습니다. 본 프로젝트에서는 좌표를 갖는 측청치를 2차원 평면 이미지로 다루기 때문에 이미지 처리에 척합한 CNN 모델에서 높은 성능을 얻을 수 있었습니다. 업무를 하며 볼 수 있는 다른 유형의 시계열, 범주형 등 데이터에도 적절한 모델을 선정하여 적용해야 원활한 문제 해결이 가능할 것 같습니다.

그리고 모델만큼 중요한 것이 데이터 세트인 것 같습니다. 실제 발생 가능한 불량 분포를 모두 잘 분류하려면 발생 가능한 모집단의 데이터 특성을 고르게 학습 데이터 세트에 반영하는 것이 필요할 것 같습니다. 초기 과정에서 데이터가 단순하고 학습 데이터와 테스트 데이터가 유사한 경우에는 머신러닝 알고리즘으로도 
높은 성능이 나타났지만 데이터를 약간 변형하자 성능이 큰 폭으로 감소하였습니다. 실제 테스트 과정에서도 깔끔하고 균일한 데이터 뿐만아니라 미세한 불량이나, 사람이 판단하기에도 애매한 불량이 존재하기 때문에 발생 가능한 모집단을 충분히 표현할 수 있는 학습 데이터를 준비하는 것이 중요하다는 생각이 들었습니다. 

반대로는, 문제의 정의에 따라 특별한 목적이 있는 경우에는 발생 가능한 많은 데이터를 반영하기보다 목적이되는 데이터만을 학습 데이터 세트로 구성하여 관련이 없는 상황에 영향을 받지 않도록 하는 방법도 고려해볼 수 있을 것 같습니다. 데이터 세트를 재구성하며 단순히 불량을 분류하는 문제에서 애매한 불량을 분류하는 문제로, 미세 불량으로 학습한 상황에서 더 필터링된 미세 불량에 대한 테스트 세트 분류로 문제가 세분화되었습니다. 이 과정에서 해결하고자 하는 문제와 그에 적절한 데이터 세트의 범위가 얼마든지 확장되고 축소될 수 있다는 사실을 알 수 있었습니다. 

한계점도 분명히 존재하는 것 같습니다. 데이터의 부족과 라벨링 문제를 고려한 데이터 생성 문제에서 수업시간에 언급된 GAN을 활용하여 기존 데이터와 유사한 라벨링 데이터를 생성해보고자 하였습니다. 하지만 프로젝트 환경에서 다량의 이미지를 데이터로 두 모델을 충분히 학습시키고, 모델 구조를 바꾸어가며 평가하기에는 시간과 하드웨어 자원 모두 부족하다고 느꼈습니다. 업무를 하면서도 서버 등 자원을 지원받을 수 있는 상황과 처리해야할 데이터의 크기 등이 매우 다르기 때문에 문제의 규모를 파악하고 진행 방향을 결정하는 것 또한 매우 중요한 일인 것 같습니다.

이러한 부분들을 종합적으로 고려하여 다시 프로젝트를 진행한다면 우선적으로 augmentation 등의 과정을 통해 데이터의 다양성을 확보하고, 이미지 데이터를 생성하여 CNN 모델을 기반으로 학습을 진행하며 더 다양한 크기와 종류의 불량에 대하여 깊이 있게 성능 개선 과정을 진행할 수 있을 것 같습니다. 그리고 augmentation 만으로 라벨링 데이터가 부족하다면 복잡한 생성 모델을 신규로 학습하는 방향 대신, 기존의 분류 모델을 활용하여 추론을 통해 불량 데이터를 새로 라벨링하고 학습에 다시 반영하는 과정으로 진행해보면 한정된 자원 안에서 더 넓은 범위의 불량을 분류할 수 있는 모델을 적용해볼 수 있을 것 같습니다.
