```python
import pandas as pd
df = pd.read_csv('../data/wine_data.csv')
features = df.columns[:-1]
X = df[features]
y = df['class']
from sklearn.model_selection import train_test_split
X_tn, X_te, y_tn, y_te = train_test_split(X, y, random_state=0)
from sklearn.preprocessing import StandardScaler
std_scale = StandardScaler()
std_scale.fit(X_tn)
X_tn_std = std_scale.transform(X_tn)
X_te_std = std_scale.transform(X_te)

from sklearn.ensemble import AdaBoostClassifier
clf = AdaBoostClassifier(random_state=0)
clf.fit(X_tn_std, y_tn)

pred = clf.predict(X_te_std)

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_te, pred)
print(accuracy)

from sklearn.metrics import precision_score
precision = precision_score(y_te, pred, average='macro')
print(precision)

from sklearn.metrics import recall_score
recall = recall_score(y_te, pred, average='macro')
print(recall)

from sklearn.metrics import f1_score
f1 = f1_score(y_te, pred, average='macro')
print(f1)

from sklearn.metrics import confusion_matrix
conf_matrix = confusion_matrix(y_te, pred)
print(conf_matrix)

from sklearn.metrics import classification_report
class_report = classification_report(y_te, pred)
print(class_report)
```

```python
import pandas as pd
df = pd.read_csv('../data/wine_data.csv')
features = df.columns[:-1]
X = df[features]
y = df['class']
from sklearn.model_selection import train_test_split
X_tn, X_te, y_tn, y_te = train_test_split(X, y, random_state=0)
from sklearn.preprocessing import StandardScaler
std_scale = StandardScaler()
std_scale.fit(X_tn)
X_tn_std = std_scale.transform(X_tn)
X_te_std = std_scale.transform(X_te)

from sklearn.ensemble import GradientBoostingClassifier
clf = GradientBoostingClassifier(max_depth=2,
                                 learning_rate=0.1,
                                 random_state=0)
clf.fit(X_tn_std, y_tn)
pred = clf.predict(X_te_std)

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_te, pred)
print(accuracy)

from sklearn.metrics import precision_score
precision = precision_score(y_te, pred, average='macro')
print(precision)

from sklearn.metrics import recall_score
recall = recall_score(y_te, pred, average='macro')
print(recall)

from sklearn.metrics import f1_score
f1 = f1_score(y_te, pred, average='macro')
print(f1)

from sklearn.metrics import confusion_matrix
conf_matrix = confusion_matrix(y_te, pred)
print(conf_matrix)

from sklearn.metrics import classification_report
class_report = classification_report(y_te, pred)
print(class_report)
```
