## - 데이터 수집 및 사용방안, 사용 데이터 정의
### &nbsp;데이터 확보 방법 : 셀의 좌표별 테스트 데이터 측정값과 값의 분포에 따른 불량 유형이 구분되어 있는 데이터를 kaggle에서 확보하였습니다.
### &nbsp;활용 데이터 정의 : wafer를 담는 용기인 lot 별 번호, wafer를 구분하는 id, wafer 내의 chip을 구분하는 x, y 좌표마다의 test 결과 측정 수치를 나타내는 데이터를 사용합니다. 측정 수치 열은 각각 불량 유형에 따라서 ring, spot, trend성 불량 등으로 구분되며 이러한 불량 유형에 해당하는 측정 값들이 해당 column에 저장되어 있습니다.

## - 분석 시 예상되는 어려움
### &nbsp;&nbsp; 1. 임의로 지정되어 있는 데이터의 저장 형태를 필요한 features와 labels의 행과 열의 형태로 변환하고, 필요에 따라 이미지로 변환하여 처리하는 전처리 과정에 어려움이 있을 것 같습니다.
### &nbsp;&nbsp;2. 불량 데이터의 분포의 일관성이나 노이즈에 따라서 모델의 성능이 크게 잘나오거나 못나오는 경우가 발생할 수 있을 것 같습니다.
### &nbsp;&nbsp;3. 데이터의 수가 많이 필요하다면 데이터가 부족할 수 있을 것 같습니다.

## - 문제 해결을 위한 모델 사용 방법
### &nbsp;&nbsp;1. 불량 유형을 구분하는 분류 문제로, 우선적으로 데이터를 vector 형태로 변환하고 주어진 label에 따라 지도학습 모델로 분류를 수행할 계획입니다.

### &nbsp;&nbsp;3. 추가적으로, 현재 프로젝트 및 현업에서의 데이터 불균형이나 부족 문제를 고려하여 주어진 측정 데이터를 기반으로 불량 유형별 데이터를 생성하는 과정도 진행해보려고 합니다.

### 1. 각 wafer 별 x, y 매핑 이미지 데이터 생성

#### 0~255 사이 min max scaling 후 값 반전하여 gray scale 값 생성
#### wafer 제품별로 측정값의 상대적인 분포에 따라 불량이 판단되므로, 
#### train 데이터 전체를 기준으로 수행하지 않고 각 wafer 단위 min max value로 scaling 수행 

```python
def scale_to_gray(data):
    min_val = np.min(data)
    max_val = np.max(data)
    scaled_data = 255 - (255 * (data - min_val) / (max_val - min_val))
    scaled_data_int = scaled_data.astype(np.uint8)
    return scaled_data_int
```

# 3) 모델 학습 및 평가

## &nbsp;전처리 데이터 data_set.csv 모델 학습 진행

### &nbsp;&nbsp;-&nbsp; 학습 및 평가 과정
    1. 전체 data map 변환 및 벡터/라벨 생성
    2. train / test 데이터 분할 및 선형 회귀 모델 학습
    3. 모델별 성능 분석

#### 1. 전체 data map 변환 및 벡터/라벨 생성

```python
df = pd.read_csv(root + '/dataset/data_set.csv', encoding='cp949')
```

```python
train_x_resize = []
train_y_resize = []
for wf_idx in range(1,101):
    df_wf = df[df['wafer']==wf_idx]
    for col in raw_col:
        map = np.zeros([116, 194])
        for index, row in df_wf.iterrows():
            map[row['x']][row['y']] = row[col]
            
        gray_map = scale_to_gray(map)
        resized_map = cv2.resize(gray_map, (15, 25))
        
        top_rows = resized_map[:5, :]   # 위에서부터 5행
        bottom_rows = resized_map[-5:, :]  # 아래에서부터 5행
        
        top_bottom = np.concatenate((top_rows, bottom_rows))
        wf_vector = top_bottom.flatten()
        
        train_x_resize.append(wf_vector)
        train_y_resize.append(col)
```

#### 2. train / test 데이터 분할 및 학습

##### &nbsp; 분류에는 회귀 모델인 Logistic Regression 사용

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(train_x_resize, train_y_resize, test_size=0.2, random_state=42)

model = LogisticRegression()
model.fit(X_train, y_train)
y_test_pred = model.predict(X_test)
test_accuracy = accuracy_score(y_test, y_test_pred)

print(f"Test Accuracy: {test_accuracy:.4f}")
```

#### 3. 모델별 성능 분석

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

X_train, X_test, y_train, y_test = train_test_split(train_x_resize, train_y_resize, test_size=0.2, random_state=42)

models = {
    "Logistic Regression": LogisticRegression(),
    "Support Vector Machine": SVC(kernel='linear'),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "k-Nearest Neighbors": KNeighborsClassifier(n_neighbors=3),
    "Multi-Layer Perceptron": MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000, random_state=42)
}

for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_test_pred = model.predict(X_test)
    test_accuracy = accuracy_score(y_test, y_test_pred)
    print(f"{model_name} Test Accuracy: {test_accuracy:.4f}")
    print()
    
    conf_matrix = confusion_matrix(y_test, y_test_pred)
    print(conf_matrix)
    print()
    
    class_report = classification_report(y_test, y_test_pred)
    print(f"{model_name} Class Report")
    print(class_report)
    print("-" * 50)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score


models = {
    "Logistic Regression": LogisticRegression(),
    "Support Vector Machine": SVC(kernel='linear'),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "k-Nearest Neighbors": KNeighborsClassifier(n_neighbors=3),
    "Multi-Layer Perceptron": MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000, random_state=42)
}

result_resize = []
data_size = [0.001 * i for i in range(600, 991)]
for size in data_size:
    X_train, X_test, y_train, y_test = train_test_split(train_x_resize, train_y_resize, test_size=size, random_state=42)
    tmp = []
    for model_name, model in models.items():
        model.fit(X_train, y_train)
        y_test_pred = model.predict(X_test)
        test_accuracy = accuracy_score(y_test, y_test_pred)
        tmp.append(test_accuracy)
    result_resize.append(tmp)
```

```python
plt.figure(figsize=(10, 6))

for i, model_name in enumerate(models.keys()):
    accuracies = [result_resize[j][i] for j in range(len(result_resize))]
    plt.plot(data_size, accuracies, label=model_name)

plt.title('Accuracy vs. Data Size')
plt.xlabel('Test Data Ratio')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score


models = {
    "Logistic Regression": LogisticRegression(),
    "Support Vector Machine": SVC(kernel='linear'),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "k-Nearest Neighbors": KNeighborsClassifier(n_neighbors=3),
    "Multi-Layer Perceptron": MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000, random_state=42)
}

result = []
data_size = [0.01 * i for i in range(1, 100)]
for size in data_size:
    X_train, X_test, y_train, y_test = train_test_split(train_x, train_y, test_size=size, random_state=42)
    tmp = []
    for model_name, model in models.items():
        model.fit(X_train, y_train)
        y_test_pred = model.predict(X_test)
        test_accuracy = accuracy_score(y_test, y_test_pred)
        tmp.append(test_accuracy)
    result.append(tmp)
```

```python
X_train = np.vstack(X_train)
X_test = np.vstack(X_test)
y_train = np.vstack(y_train)
y_test = np.vstack(y_test)
```

```python
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Input, Concatenate
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder


latent_dim = 100
input_dim = X_train.shape[1]

num_classes = 5

label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_train_one_hot = to_categorical(y_train_encoded, num_classes=num_classes)

# Generator 모델 정의
gen_input = Input(shape=(latent_dim,))
gen_label = Input(shape=(num_classes,))
gen_concat = Concatenate()([gen_input, gen_label])
gen = Dense(128, activation='relu')(gen_concat)
gen = Dense(256, activation='relu')(gen)
gen = Dense(512, activation='relu')(gen)
gen_output = Dense(input_dim, activation='sigmoid')(gen)
generator = Model([gen_input, gen_label], gen_output)

# Discriminator 모델 정의
disc_input = Input(shape=(input_dim,))
disc_label = Input(shape=(num_classes,))
disc_concat = Concatenate()([disc_input, disc_label])
disc = Dense(512, activation='relu')(disc_concat)
disc = Dense(256, activation='relu')(disc)
disc = Dense(128, activation='relu')(disc)
disc_output = Dense(1, activation='sigmoid')(disc)
discriminator = Model([disc_input, disc_label], disc_output)

# 컴파일 Discriminator
discriminator.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['accuracy'])
discriminator.trainable = False

# GAN 모델 정의
gan_input = Input(shape=(latent_dim,))
gan_label = Input(shape=(num_classes,))
gan_gen_output = generator([gan_input, gan_label])
gan_disc_output = discriminator([gan_gen_output, gan_label])
gan = Model([gan_input, gan_label], gan_disc_output)
gan.compile(optimizer=Adam(), loss='binary_crossentropy')

# cGAN 모델 학습
epochs = 100
batch_size = 32
for epoch in range(epochs):
    noise = np.random.randn(batch_size, latent_dim)
    sampled_labels = np.random.randint(0, num_classes, batch_size)
    sampled_labels_one_hot = to_categorical(sampled_labels, num_classes=num_classes)
    generated_data = generator.predict([noise, sampled_labels_one_hot])

    real_data = X_train[np.random.randint(0, X_train.shape[0], batch_size)]
    real_labels = y_train_one_hot[np.random.randint(0, y_train_one_hot.shape[0], batch_size)]

    # Discriminator 훈련
    X = np.concatenate([real_data, generated_data])
    y = np.concatenate([np.ones((batch_size, 1)), np.zeros((batch_size, 1))])
    labels = np.concatenate([real_labels, sampled_labels_one_hot])
    d_loss = discriminator.train_on_batch([X, labels], y)

    # Generator 훈련
    noise = np.random.randn(batch_size, latent_dim)
    y_gen = np.ones((batch_size, 1))
    g_loss = gan.train_on_batch([noise, sampled_labels_one_hot], y_gen)

    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Discriminator Loss: {d_loss[0]}, Generator Loss: {g_loss}")



```

```python
# 학습된 Generator를 사용하여 특정 클래스의 데이터 생성
def generate_data(generator, latent_dim, class_label, num_samples=1):
    noise = np.random.randn(num_samples, latent_dim)
    label_one_hot = to_categorical([class_label] * num_samples, num_classes=num_classes)
    generated_data = generator.predict([noise, label_one_hot])
    return generated_data * 255.0
```
