```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=200, n_features=2, centers=2, random_state=2)
```

```python
print(X.shape)
print(y.shape)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0], X[:,1], c='gray', edgecolor='black', marker='o')
plt.show()
```

```python
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(n_components=2, random_state=0)
gmm.fit(X)
label_gmm = gmm.predict(X)
print(label_gmm)
```

```python
import pandas as pd
gmm_columns = ['gmm_comp1', 'gmm_comp2']
X_gmm_df = pd.DataFrame(X, columns=gmm_columns)
X_gmm_df['target'] = y
X_gmm_df['label_gmm'] = label_gmm
X_gmm_df.head(5)
```

```python
df = X_gmm_df
markers=['o', 'x']

for i, mark in enumerate(markers):
    df_i = df[df['label_gmm']==i]
    target_i = i
    X1 = df_i['gmm_comp1']
    X2 = df_i['gmm_comp2']
    plt.scatter(X1, X2, marker=mark, label=target_i)

plt.xlabel('gmm_component1')
plt.ylabel('gmm_component2')
plt.legend()
plt.show()
```

```python
df = X_gmm_df
markers=['o', 'x']

for i, mark in enumerate(markers):
    df_i = df[df['target']==i]
    target_i = i
    X1 = df_i['gmm_comp1']
    X2 = df_i['gmm_comp2']
    plt.scatter(X1, X2, marker=mark, label=target_i)

plt.xlabel('gmm_component1')
plt.ylabel('gmm_component2')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score
sil_score = silhouette_score(X, label_gmm)
print(sil_score)
```

```python
from sklearn.preprocessing import StandardScaler
std_scale = StandardScaler()
std_scale.fit(X_tn)
X_tn_std = std_scale.transform(X_tn)
```

```python
from sklearn.preprocessing import StandardScaler
std_scale = StandardScaler()
std_scale.fit(X_tn)
X_tn_std = std_scale.transform(X_tn)
X_te_std = std_scale.transform(X_te)
```

```python
pca_columns = ['pca_comp1', 'pca_comp2']
X_tn_pca_df = pd.DataFrame(X_tn_pca, columns=pca_columns)
y_tn1 =  y_tn.reset_index(drop=True)
X_tn_pca_df['target'] = y_tn1
X_tn_pca_df.head(10)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,8))
plt.scatter(X_tn_pca_df['pca_comp1'],
            X_tn_pca_df['pca_comp2'],
            marker='o')
plt.xlabel('pca_component1')
plt.ylabel('pca_component2')
plt.show()
```

```python
df = X_tn_pca_df
df_0 = df[df['target']==0]
df_1 = df[df['target']==1]
df_2 = df[df['target']==2]

X_11 = df_0['pca_comp1']
X_12 = df_1['pca_comp1']
X_13 = df_2['pca_comp1']

X_21 = df_0['pca_comp2']
X_22 = df_1['pca_comp2']

target_0 = 0
target_1 = 1
target_2 = 2
X_23 = df_2['pca_comp2']

plt.figure(figsize=(15,8))
plt.scatter(X_11, X_21, marker='o', label=target_0)
plt.scatter(X_12, X_22, marker='x', label=target_1)
plt.scatter(X_13, X_23, marker='^', label=target_2)

plt.xlabel('pca_component1')
plt.ylabel('pca_component2')
plt.legend()
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf_knn = KNeighborsClassifier(n_neighbors=2)
clf_knn.fit(X_tn_pca, y_tn)
```

```python
knn_pred = clf_knn.predict(X_te_pca)
print(knn_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_te, knn_pred)
print(accuracy)
```

```python
import pandas as pd
df = pd.read_csv('../data/wine_data.csv')
```

```python
features = df.columns[:-1]
X = df[features]
y = df['class']
```

```python
from sklearn.model_selection import train_test_split
X_tn, X_te, y_tn, y_te = train_test_split(X, y, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
std_scale = StandardScaler()
std_scale.fit(X_tn)
X_tn_std = std_scale.transform(X_tn)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis()
lda.fit(X_tn_std, y_tn)
```

```python
pred_lda = lda.predict(X_te_std)
```

```python
from sklearn.metrics import accuracy_score
accuracy_lda = accuracy_score(y_te, pred_lda)
print(accuracy_lda)
```

```python
X_tn_lda = lda.transform(X_tn_std)
X_te_lda = lda.transform(X_te_std)
```

```python
print(X_tn_std.shape)
print(X_tn_lda.shape)
```

```python
print(lda.intercept_)
```

```python
print(lda.coef_)
```

```python
lda_columns = ['lda_comp1', 'lda_comp2']
X_tn_lda_df = pd.DataFrame(X_tn_lda, columns=lda_columns)
X_tn_lda_df['target'] = y_tn.reset_index(drop=True)
X_tn_lda_df.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
clf_rf_lda = RandomForestClassifier(max_depth=2, random_state=0)
clf_rf_lda.fit(X_tn_lda, y_tn)
pred_rf_lda = clf_rf_lda.predict(X_te_lda)
```

```python
from sklearn.metrics import accuracy_score
accuracy_lda = accuracy_score(y_te, pred_rf_lda)
print(accuracy_lda)
```
