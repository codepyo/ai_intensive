```python
import numpy as np
import pandas as pd
import statsmodels
from statsmodels.regression import linear_model
from statsmodels.tsa.arima.model import ARIMA
from arch import arch_model
```

```python
df = pd.read_csv('../data/air_quality_uci.csv', encoding='cp949')
df.head(3)
```

```python
df.shape
```

```python
df['DateTime'] = df['Date'] + ' ' + df['Time']
df.head(3)
```

```python
df['DateTime'][3]
```

```python
type(df['DateTime'][3])
```

```python
df['DateTimeIndex'] = pd.to_datetime(df['DateTime'])
df.head(3)
```

```python
type(df['DateTimeIndex'][3])
```

```python
df = df.set_index(keys=['DateTimeIndex'])
```

```python
df.head(3)
```

```python
df.isnull().sum()
```

```python
train_ratio = 0.9
n = len(df)
n_train = int(n*train_ratio)
df_train = df[:n_train]
df_test = df[n_train:]
```

```python
df.columns
```

```python
features = ['T', 'AH']
target = 'PT08_S4'

X_tn = df_train[features]
y_tn = df_train[target]
X_te = df_test[features]
y_te = df_test[target]
```

```python
X_tn
```

```python
ols_reg = linear_model.OLS(endog=y_tn, exog=X_tn)
ols_model = ols_reg.fit()
```

```python
ols_model.params
```

```python
ols_model.summary()
```

```python
ols_model.predict()
```
