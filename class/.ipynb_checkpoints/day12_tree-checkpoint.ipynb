```python
import pandas as pd
df = pd.read_csv('../data/wine_data.csv')
```

```python
features = df.columns[:-1]
X = df[features]
y = df['class']
```

```python
from sklearn.model_selection import train_test_split
X_tn, X_te, y_tn, y_te = train_test_split(X, y, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
std_scale = StandardScaler()
std_scale.fit(X_tn)
X_tn_std = std_scale.transform(X_tn)
X_te_std = std_scale.transform(X_te)
```

```python
from sklearn import tree
clf_tree = tree.DecisionTreeClassifier(random_state=0)
clf_tree.fit(X_tn_std, y_tn)
```

```python
pred_tree = clf_tree.predict(X_te_std)
print(pred_tree)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_te, pred_tree)
print(accuracy)
```

```python
from sklearn.metrics import precision_score
precision = precision_score(y_te, pred_tree, average='macro')
print(precision)
```

```python
from sklearn.metrics import recall_score
recall = recall_score(y_te, pred_tree, average='macro')
print(recall)
```

```python
from sklearn.metrics import f1_score
f1 = f1_score(y_te, pred_tree, average='macro')
print(f1)
```

```python
from sklearn.metrics import confusion_matrix
conf_matrix = confusion_matrix(y_te, pred_tree)
print(conf_matrix)
```

```python
from sklearn.metrics import classification_report
class_report = classification_report(y_te, pred_tree)
print(class_report)
```

```python
df = pd.read_csv('../data/house_prices.csv')
```

```python
df
```

```python
df['MEDV'].unique()
```

```python
features = df.columns[:-1]
X = df[features]
y = df['MEDV']
```

```python
from sklearn.model_selection import train_test_split
X_tn, X_te, y_tn, y_te = train_test_split(X, y, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
std_scale = StandardScaler()
std_scale.fit(X_tn)
X_tn_std = std_scale.transform(X_tn)
X_te_std = std_scale.transform(X_te)
```

```python
from sklearn.tree import DecisionTreeRegressor
clf_tree = DecisionTreeRegressor(random_state=0)
clf_tree.fit(X_tn_std, y_tn)
```

```python
tree_pred = clf_tree.predict(X_te_std)
print(tree_pred)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_te, tree_pred)
print(mse)
```
